```python
import akshare as ak
import pandas as pd
import time
import os

from tenacity import retry, stop_after_attempt, wait_fixed
```

```python
sh_stock_list_df = pd.read_csv('../data/stock/list/sh_stock_list.csv', dtype={'code': str})
sz_stock_list_df = pd.read_csv('../data/stock/list/sz_stock_list.csv', dtype={'code': str})
```

```python
count = 0
for index, row in sz_stock_list_df.iterrows():
    stock_code = row['A股代码']
    stock_name = row['A股简称']
    # print(f"Downloading data for {stock_code} - {stock_name}")
    if row['板块'] != '主板':
        print(f"Skipping {stock_code} - {stock_name} due to not being main board")
        continue
    if "ST" in stock_name:
        print(f"Skipping {stock_code} - {stock_name} due to ST status")
        continue

    count += 1
    
print(f"Total stocks to download from SH: {count}")
```

```python
@retry(stop=stop_after_attempt(5), wait=wait_fixed(2))
def download_stock_data(stock_code, start_date, end_date):
    try:
        # stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=stock_code, period="daily", start_date=start_date, end_date=end_date, adjust="hfq")
        stock_zh_a_daily_df = ak.stock_zh_a_daily(symbol=stock_code, start_date=start_date, end_date=end_date, adjust="hfq")
        
        # return stock_zh_a_hist_df
        return stock_zh_a_daily_df
    except Exception as e:
        print(f"Error downloading data for {stock_code}: {e}, retrying...")
        raise
```

```python
data_dir = "/home/zzy/projects/stocka/data/stock/kline/day"
start_date = '20050101'
end_date = '20260202'
```

```python
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

sh_count = 0
for index, row in sh_stock_list_df.iterrows():
    stock_code = str(row['证券代码'])
    stock_name = row['证券简称']
    # print(f"Downloading data for {stock_code} - {stock_name}")
    if "ST" in stock_name:
        print(f"Skipping {stock_code} - {stock_name} due to ST status")
        continue
    # if stock_code != "601599":
    #     continue
    stock_zh_a_hist_df = download_stock_data(f"sh{stock_code}", start_date, end_date)
    stock_file_name = os.path.join(data_dir, f'{stock_code}.csv')
    stock_zh_a_hist_df.to_csv(stock_file_name, index=False)
    sh_count += 1
    time.sleep(1)  # To avoid hitting rate limits
    # break
print(f"Total stocks downloaded from SH: {sh_count}")
```

```python
sz_count = 0
for index, row in sz_stock_list_df.iterrows():
    stock_code = f"{row['A股代码']:06d}"
    stock_name = row['A股简称']
    # print(f"Downloading data for {stock_code} - {stock_name}")
    if row['板块'] != '主板':
        print(f"Skipping {stock_code} - {stock_name} due to not being main board")
        continue
    if "ST" in stock_name:
        print(f"Skipping {stock_code} - {stock_name} due to ST status")
        continue
    
    stock_zh_a_hist_df = download_stock_data(f"sz{stock_code}", start_date, end_date)
    stock_file_name = os.path.join(data_dir, f'{stock_code}.csv')
    stock_zh_a_hist_df.to_csv(stock_file_name, index=False)
    sz_count += 1
    time.sleep(1)  # To avoid hitting rate limits
print(f"Total stocks downloaded from SZ: {sz_count}")
```
